# Null hypothesis1:
## H0: There is no statistical difference in the age distribution of male and female riders
## H1: There is statistical difference in the age distribution of male and female riders
## $$ \alpha = 0.05 $$

```python
df.index=range(len(df.tripduration))
```

```python
df['ageM'] = 2015 - df['birth year'][(df['usertype'] == 'Subscriber') & (df['gender'] == 1)]
df['ageF'] = 2015 - df['birth year'][(df['usertype'] == 'Subscriber') & (df['gender'] == 2)]
```

```python
df.describe()
```

```python
df['ageM'].dropna(inplace= True)
df['ageF'].dropna(inplace= True)
df.head(20)
```

```python
bins = np.arange(10, 99, 5)
axM = df.ageM.groupby(pd.cut(df.ageM, bins)).agg([count_nonzero]).plot(kind='bar', 
                                                                legend=False)
axM.set_title("male riders")
axF = df.ageF.groupby(pd.cut(df.ageF, bins)).agg([count_nonzero]).plot(kind='bar',
                                                                legend=False)
axF.set_title("female riders")
```

```python
csM=df.ageM.groupby(pd.cut(df.ageM, bins)).agg([count_nonzero]).cumsum()

csF=df.ageF.groupby(pd.cut(df.ageF, bins)).agg([count_nonzero]).cumsum()

print (np.abs(csM / csM.max()-csF / csF.max()))

pl.plot(bins[:-1] + 5, csM / csM.max(), label = "M")
pl.plot(bins[:-1] + 5, csF / csF.max(), label = "F")
pl.plot(bins[:-1] + 5, np.sqrt(csF / csF.max() - csM / csM.max())**2, 'k-',
        label = "difference")
pl.xlabel("Age")
pl.ylabel("Normalized Cumulative Number")
pl.legend()
```

```python
import scipy.stats
```

# KS: alpha = 0.05
## entire dataset:

```python
ks = scipy.stats.ks_2samp(df.ageM, df.ageF)
```

```python
print (ks)
```

## For the entire dataset, since the p-value=0.0 < 0.05, we reject the H0: There is no statistical difference in the age distribution of male and female riders
## subset 200 times smaller:

```python
subdf = df.loc[::200,:]
```

```python
subdf['ageM'] = 2015 - subdf['birth year'][(subdf['usertype'] == 'Subscriber') & (subdf['gender'] == 1)]
subdf['ageF'] = 2015 - subdf['birth year'][(subdf['usertype'] == 'Subscriber') & (subdf['gender'] == 2)]
```

## For the 200 times smaller dataset, since the p-value=0.0 < 0.05, we reject the H0: There is no statistical difference in the age distribution of male and female riders

# Pearson's test for correlation:

```python
MD = df.ageM.groupby(pd.cut(df.ageM, bins)).agg([count_nonzero])
FD = df.ageF.groupby(pd.cut(df.ageF, bins)).agg([count_nonzero])

MD = list(MD.count_nonzero) 
FD = list(FD.count_nonzero) 

MD = MD / np.sum(MD)
FD = FD / np.sum(FD)

scipy.stats.pearsonr(MD,FD)
```

```python
pvalue = 2.4715305348508075e-11
```

```python
pvalue < 0.05
```

## The correlation is 0.9728 which means that there is a strong positive correlation between the age distribution of male and female. Since p-value=2.4715305348508075e-11 < 0.05, we reject the H0: There is no statistical difference in the age distribution of male and female riders

# Spearman's test for correlation:

```python
scipy.stats.spearmanr(MD,FD)
```

```python
pvalue = 4.696736001555885e-63
pvalue < 0.05
```

## The correlation is 0.9810 which means that there is a strong positive correlation between the age distribution of male and female. Since p-value=4.696736001555885e-63 < 0.05, we reject the H0: There is no statistical difference in the age distribution of male and female riders

# Null hypothesis2:
